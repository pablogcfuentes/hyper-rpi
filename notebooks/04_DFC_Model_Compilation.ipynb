# Fase 04: Compilación desde el Hailo Archive Quantized Model a HEF (Hailo Executable Format)

```python
# Imports

from hailo_sdk_client import ClientRunner
```

## 1. Compilar el modelo HAR cuantizado a modelo HEF ejecutable

```python
# Cargamos el modelo HAR cuantizado

model_name = "../results/model_cnn21"
quantized_model_har_path = f"{model_name}_quantized_model.har"
```

```python
# Cargar el ClientRunner

runner = ClientRunner(har=quantized_model_har_path)
# By default it uses the hw_arch that is saved on the HAR. It is not recommended to change the hw_arch after Optimization.
```

```python
# Ejecutar el compilador

hef = runner.compile()

file_name = f"{model_name}.hef"
with open(file_name, "wb") as f:
    f.write(hef)
```

### 1.1. Mostrar información sobre el modelo

```python
# Guardar el HAR compilado y mostrar información del modelo

har_path = f"{model_name}_compiled_model.har"
runner.save_har(har_path)
!hailo profiler {har_path}
```
