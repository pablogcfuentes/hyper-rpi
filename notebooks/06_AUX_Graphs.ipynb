# Fase 06: Generar gráficas de los resultados

```python
# Imports

import matplotlib.pyplot as plt
import numpy as np
```

## 1. Generar gráficas para OA, AA y tiempo de inferencia

```python
# Datos iniciales

model_names = [
    "base",
    "o0_c0",
    "o1_c0",
    "o2_c0",
    "o4_c0",
    "o4_c1",
]
```

```python
# Graficar los tiempos de inferencia
inference_times = [
    503.0986,
    666.1553,
    680.5718,
    671.5710,
    665.2532,
    667.3588
]

plt.figure(figsize=(8, 4))
bars = plt.bar(model_names, inference_times, color='skyblue')

# Añadimos valores encima de las barras
for bar, value in zip(bars, inference_times):
    plt.text(bar.get_x() + bar.get_width() / 2,
             bar.get_height() + 2,
             f'{value: .2f}',
             ha='center', va='bottom', fontsize=10)

plt.ylim(top=max(inference_times) * 1.1)
plt.ylabel("Inference Time (s)")
plt.title("Inference Times for model_cnn21 with different optimizations and compressions")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Guardar y mostrar la imagen
plt.savefig("../results/graphs/inference_times.png")
plt.savefig("../results/graphs/inference_times.pdf", dpi=300)
plt.show()
```

```python
# Graficar los OA
oa_values = [
    87.95,
    86.93,
    86.17,
    4.63,
    87.13,
    87.02
]

plt.figure(figsize=(8, 4))
bars = plt.bar(model_names, oa_values, color='skyblue')

# Añadimos valores encima de las barras
for bar, value in zip(bars, oa_values):
    plt.text(bar.get_x() + bar.get_width() / 2,
             bar.get_height() + 2,
             f'{value: .2f}',
             ha='center', va='bottom', fontsize=10)

plt.ylim(top=max(oa_values) * 1.1)
plt.ylabel("Overall Acuraccy (%)")
plt.title("Overall Accuracies for model_cnn21 with different optimizations and compressions")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Guardar y mostrar la imagen
plt.savefig("../results/graphs/overall_accuracies.png")
plt.savefig("../results/graphs/overall_accuracies.png", dpi=300)
plt.show()
```

```python
# Graficar los AA
aa_values = [
    85.88,
    84.53,
    82.82,
    10.58,
    83.28,
    83.99
]

plt.figure(figsize=(8, 4))
bars = plt.bar(model_names, aa_values, color='skyblue')

# Añadimos valores encima de las barras
for bar, value in zip(bars, aa_values):
    plt.text(bar.get_x() + bar.get_width() / 2,
             bar.get_height() + 2,
             f'{value: .2f}',
             ha='center', va='bottom', fontsize=10)

plt.ylim(top=max(aa_values) * 1.1)
plt.ylabel("Average Acuraccy (%)")
plt.title("Average Accuracies for model_cnn21 with different optimizations and compressions")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Guardar y mostrar la imagen
plt.savefig("../results/graphs/average_accuracies.png")
plt.savefig("../results/graphs/average_accuracies.png", dpi=300)
plt.show()
```

```python
# Graficar los OA y los AA

# Posiciones en el eje x
x = np.arange(len(model_names))
width = 0.35 # Ancho de cada barra

plt.figure(figsize=(8, 4))

# Dibujamos las barras OA y AA con desplazamiento
bars1 = plt.bar(x - width/2, oa_values, width, label='OA', color='skyblue')
bars2 = plt.bar(x + width/2, aa_values, width, label='AA', color='orange')

# Añadimos valores encima de las barras
for bar, value in zip(bars1, oa_values):
    plt.text(bar.get_x() + bar.get_width() / 2,
             bar.get_height() + 2,
             f'{value: .2f}',
             ha='center', va='bottom', fontsize=10)
    
for bar, value in zip(bars2, aa_values):
    plt.text(bar.get_x() + bar.get_width() / 2,
             bar.get_height() + 2,
             f'{value: .2f}',
             ha='center', va='bottom', fontsize=10)

plt.ylim(top=max(max(oa_values), max(aa_values)) * 1.1)
plt.ylabel("Accuracy (%)")
plt.title("OA vs AA for model_cnn21 with different optimizations and compressions")
plt.xticks(x, model_names, rotation=45)
plt.legend(loc='lower right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Guardar y mostrar la imagen
plt.savefig("../results/graphs/oa_vs_aa.png")
plt.savefig("../results/graphs/oa_vs_aa.pdf", dpi=300)
plt.show()
```
